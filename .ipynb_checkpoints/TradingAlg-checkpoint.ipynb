```python
import pandas as pd
import numpy as np
import matplotlib
```

# Preprocessing Data

## Import Data

```python
pd.set_option('max_columns', None)
```

### Securities dataset

```python
securities = pd.read_csv("./nyse/securities.csv")
securities.head(5)
```

```python
# get info on the securities data 
num_securities = len(securities) # get the number of securities 
gics_sector_values = np.unique(securities['GICS Sector']) # get the unique GICS Sector Values 

print('The number of securities is ', num_securities)
print('The GICS Sectors are ', gics_sector_values)
```

```python
# get the rows where the value in GICS Sector column is 'Information Technology'
it_securities = securities.loc[securities['GICS Sector'] == 'Information Technology'] # get the IT securities
num_it_securities = len(it_securities) # get the number of IT securities 

print('The number of IT securities is ', num_it_securities)
print('The ticker symbols of the IT securities are ', it_securities['Ticker symbol'])
```

```python
# select the securities for Yahoo and Google 

yahoo_sec = securities.loc[securities['Ticker symbol'] == "YHOO"]
google_sec = securities.loc[securities['Ticker symbol'] == "GOOG"]

print('Yahoo Security: \n', yahoo_sec)
print('Google Security: \n', google_sec)
```

### Prices dataset

```python
# prices dataset 
prices = pd.read_csv("./nyse/prices.csv", parse_dates=['date'])
prices.head(5)
```

```python
# select all the stocks with ticker values 'GOOG' and set the index to date
# https://www.kaggle.com/residentmario/time-series-plotting-optional
google_prices = prices.loc[prices['symbol'] == "GOOG"].set_index('date')
google_prices.head(5)
```

```python
# do the same for yahoo
yahoo_prices = prices.loc[prices['symbol'] == "YHOO"].set_index('date')
yahoo_prices.head(5)
```

```python
# visualize the Google stock data by year
# https://www.kaggle.com/residentmario/time-series-plotting-optional
google_prices['volume'].resample('Y').mean().plot.bar()
```

```python
# do the same for Yahoo
yahoo_prices['volume'].resample('Y').mean().plot.bar()
```

### Fundamentals dataset 

```python
fundamentals = pd.read_csv("./nyse/fundamentals.csv").set_index('Period Ending')
fundamentals.head(5)
```

```python
# select all the Google fundamentals
google_fundamentals = fundamentals.loc[fundamentals['Ticker Symbol'] == "GOOG"]
google_fundamentals.head(5)
```

```python
# select all the Yahoo fundamentals
yahoo_fundamentals = fundamentals.loc[fundamentals['Ticker Symbol'] == "YHOO"]
yahoo_fundamentals.head(5)
```

### Prices split adjusted dataset

```python
prices_adj = pd.read_csv("./nyse/prices-split-adjusted.csv", parse_dates=['date'])
prices_adj.head(5)
```

```python
# select all the Google stocks adjusted
google_prices_adj = prices_adj.loc[prices_adj['symbol'] == "GOOG"].set_index('date')
google_prices_adj.head(5)
```

```python
# select all the Yahoo stocks adjusted 
yahoo_prices_adj = prices_adj.loc[prices_adj['symbol'] == "YHOO"].set_index('date')
yahoo_prices_adj.head(5)
```
